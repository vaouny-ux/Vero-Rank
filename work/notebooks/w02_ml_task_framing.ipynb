## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

The task type is Classification. Specifically, it is a binary or multi-class classification problem where we predict whether a given keyword/page will achieve a top-10 ranking (High Visibility) or classify the "Search Intent" (Informational, Transactional, Navigational) of a keyword based on SERP features and metrics. This helps prioritize SEO content production effectively.

```python
print("Task Type: Classification")
print("Target Type: Categorical / Discrete Labels")
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

The target variable is an observed outcome: whether a page/keyword successfully ranks in the Top 10 of Google search results (e.g., is_top_10 encoded as 1 or 0). The label comes from historical performance data or tracked rank positions from SEO tools (like Google Search Console or Ahrefs) over a specific observation window, serving as a clean, measured proxy for "high SEO value."

```python
print("Target column structure: Binary (0 = No, 1 = Yes)")
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

The chosen success metric is Precision paired with ROC-AUC. In SEO optimization, content creation is expensive. We want to be highly confident that when the model predicts a keyword will break into the Top 10, it actually does (minimizing False Positives). A Precision of > 80% and an ROC-AUC of > 0.85 would mean decision-support we can trust.

```python
from sklearn.metrics import precision_score, roc_auc_score
print("Target Metric for Business Decision: Precision > 0.80")
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

SEO algorithms are highly dynamic and non-linear. A simple fixed rule like if search_volume > 5000 and difficulty < 20 then target = 1 fails in reality. Google's ranking factors involve complex interactions between domain authority, content depth, search intent shifts, and user behavior. ML beats fixed rules because it can detect directional, non-linear patterns across dozens of features simultaneously, adapting as search behaviors evolve.



```python
print("ML advantage: Learns complex decision boundaries instead of hard-coded thresholds.")
```
