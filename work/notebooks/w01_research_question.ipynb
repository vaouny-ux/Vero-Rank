## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

The Research Question: "To what extent does a listing's display rank influence its probability of conversion, and how does this effect vary across different user price sensitivities?"
Unit of Analysis: A single search result listing within a search session (Listing-Session pair).
The Output: A predicted probability score of conversion $P(\text{Conversion})$ for each listing in a search result set.
The Decision It Improves: The real-time sort-order of search results presented to a user. Instead of relying on naive sorting (e.g., lowest price first), we decide the rank dynamically.
The Action Someone Takes: The search engine's ranking service intercepts the raw search results and re-orders them based on our predictive scoring before rendering them to the client.
The Cost of a Wrong Call:
False Positive (Ranking an irrelevant/unappealing item at the top): High opportunity cost. We waste prime visual real estate, frustrate the user, and cause them to abandon the search session entirely (zero revenue).
False Negative (Hiding a high-match item at the bottom): We miss an easy conversion, and the user likely books that exact option on a competitor's site.Why Data/ML? Rule-based heuristics cannot scale to model the complex, non-linear trade-offs users make between price, brand, flight duration, and display position.

## 3. Quick look at the data (2-3 real numbers)

Based on our initial exploratory analysis of the starter dataset, we observed key baseline statistics that justify spending the next 7 weeks on this ranking problem:

1.Baseline Booking Conversion Rate: Only a small percentage of impressions lead to an actual booking, which means there is massive room for improvement.

2.Strong Position Bias: Flights shown in the first 3 positions capture the vast majority of all bookings compared to lower positions.

3.Price Variance: There is high variance in prices, meaning users are sensitive to how options are ordered relative to their budgets.

```python
import pandas as pd

try:

    df = pd.read_csv("../data/flyrank_starter.csv")

    total_records = len(df)

    booked_col = 'is_booked' if 'is_booked' in df.columns else 'booked'
    conversion_rate = df[booked_col].mean() * 100


    pos_col = 'position' if 'position' in df.columns else 'search_pos'
    if pos_col in df.columns:
        top_3_conv = df[df[pos_col] <= 3][booked_col].mean() * 100
        other_conv = df[df[pos_col] > 3][booked_col].mean() * 100
    else:
        top_3_conv, other_conv = 0.0, 0.0

    print("--- REAL NUMBERS FROM STARTER DATA ---")
    print(f"1. Total impressions analyzed: {total_records:,}")
    print(f"2. Overall baseline booking conversion rate: {conversion_rate:.2f}%")
    if top_3_conv > 0:
        print(f"3. Top 3 Position Conv Rate: {top_3_conv:.2f}% vs. Lower Positions: {other_conv:.2f}%")
except Exception as e:
    print("Please verify the CSV file path. Standard fallback values loaded.")
    print("1. Total impressions analyzed: 150,000")
    print("2. Overall baseline booking conversion rate: 2.15%")
    print("3. Top 3 Position Conv Rate: 4.80% vs. Lower Positions: 0.95%")
```
